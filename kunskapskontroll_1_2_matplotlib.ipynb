### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python

import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()

# Matplotlib tvo ways

#Objektorienterad stil - rekomenderad 

# Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style").
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.

#Pyplot stil
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
import matplotlib.pyplot as plt

#Figure
#Figure är det övergripande fönstret eller containern för allt man ritar.
#Axes
#Axes är själva plotområdet inom figuren, där data visualiseras. Varje plot eller graf har sina egna Axes, och man kan ha flera av dem i en enda Figure.
#Axis
#Axis är de enskilda axlarna inom ett Axes-objekt, som x-axeln och y-axeln. Dessa visar tickmarkeringar och etiketter för att hjälpa till med att tolka grafens data.
#Artist
#Artist är alla visuella element i Matplotlib, från linjer och texter till punkter och former. Det är byggstenarna som utgör plottarna och läggs till inom Axes.

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()

import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(4, 2.5))
circle = mpatches.Circle((0.5, 0.5), 0.25, ec="none")
ax.add_artist(circle)
clipped_circle = mpatches.Circle((1, 0.5), 0.125, ec="none", facecolor='C1')
ax.add_artist(clipped_circle)
ax.set_aspect(1)
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#pyplot-metoden

import numpy as np
import matplotlib.pyplot as plt

# Skapa data
x = np.linspace(-4, 4, 100)
y = x**2

# Skapa plott med pyplot
plt.plot(x, y)
plt.title("Plot of $y = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# Objekttorienterad metod
# Skapa figuren och axlarna
fig, ax = plt.subplots()

# Plotta data
ax.plot(x, y)
ax.set_title("Plot of $y = x^2$")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Visa grafen
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Skapa data för scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Skapa data för bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Skapa figuren och två subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Första subplotten: Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, color='blue')
ax1.set_title("Scatter Plot")
ax1.set_xlabel("Random X")
ax1.set_ylabel("Random Y")

# Andra subplotten: Bar plot
ax2.bar(names, values, color='orange')
ax2.set_title("Bar Plot")
ax2.set_xlabel("Fruits")
ax2.set_ylabel("Quantity")

# Justera layout och visa plot
plt.tight_layout()
plt.show()
```
